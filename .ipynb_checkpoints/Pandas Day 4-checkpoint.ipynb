```python
import pandas as pd
df_emp = pd.read_csv("employees.csv")
df_sal = pd.read_csv("salary.csv")
```

```python
df_sal
```

```python
df_emp
```

```python
# Inner join where common key is different name
# SQL : select e.eno,e.ename,s.sal from employees as e inner join salary as s on e.eid = s.eno
df_emp.merge(df_sal,left_on='eid',right_on='eno',how='inner')
```

```python
# Left outer join where common key is different name
# SQL : select e.eno,e.ename,s.sal from employees as e left join salary as s on e.eid = s.eno
df_emp.merge(df_sal,left_on='eid',right_on='eno',how='left')
```

```python
# right outer join where common key is different name
# SQL : select e.eno,e.ename,s.sal from employees as e right join salary as s on e.eid = s.eno
df_emp.merge(df_sal,left_on='eid',right_on='eno',how='right')
```

```python
# Full outer join where common key is different name
# SQL ( Oracle ) : select e.eno,e.ename,s.sal from employees as e outer join salary as s on e.eid = s.eno
# MYSQL : 
select e.eno,e.ename,s.sal from employees as e left join salary as s on e.eid = s.eno
union
select e.eno,e.ename,s.sal from employees as e right join salary as s on e.eid = s.eno

df_emp.merge(df_sal,left_on='eid',right_on='eno',how='outer')
```

# Concatenation

```python
df_sal
```

```python
df_emp
```

```python
# Vertical concatenation
pd.concat([df_emp,df_sal],axis=0)
```

```python
# Horizontal concatenation
pd.concat([df_emp,df_sal],axis=1)
```

```python
# Vertical concatenation ( intersection of columns - common column name)
pd.concat([df_emp,df_sal],axis=0,join='inner',ignore_index=False)
```

```python
# Vertical concatenation ( union of columns )
pd.concat([df_emp,df_sal],axis=0,join='outer')
```
